```python
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()

# 데이터와 타겟 데이터를 각각 판다스의 데이터프레임으로 변환
boston_df_data = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df_target = pd.DataFrame(boston.target, columns=["MEDV"])

boston_df = pd.concat([boston_df_data, boston_df_target], axis=1)
```

```python
boston_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', context='notebook')
cols = ['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']
sns.pairplot(boston_df[cols], size=2.5)
plt.show()
sns.reset_orig()
```

```python
# 데이터 간의 상관관계
# 히트맵(heat map): 양의 상관도가 높으면 빨간색, 음의 상관도가 높으면 파란색
import numpy as np

plt.figure(figsize=(12, 9))
cm = np.corrcoef(boston_df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f',
                     annot_kws={'size': 18}, linewidths=0.5,
                     yticklabels=cols, xticklabels=cols)

plt.show()
```

```python
import matplotlib
from matplotlib import style
from sklearn.linear_model import LinearRegression

style.use('seaborn-talk')

X = boston_df[['RM']].values
y = boston_df['MEDV'].values

lin_reg = LinearRegression()
lin_reg.fit(X, y)

slope = lin_reg.coef_[0]
intercept = lin_reg.intercept_

print('회귀선 기울기: %.3f\n절편: %.3f' %(slope, intercept))

plt.scatter(X, y, c='b')
plt.plot(X, lin_reg.predict(X), c='r')

plt.xlabel('평균 방수[RM]')
plt.ylabel('1,000달러 단위 집값 [MEDV]')
plt.title('RM-MEDV predict')
plt.show()
```

```python
# 방 수가 5개인 경우 예측되는 집값 구하기
predict_rooms = [[5.0]]
house_val = lin_reg.predict(predict_rooms)

print('%.f dollar' %(house_val * 1000))
```
